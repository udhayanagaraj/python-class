Creating an dict

```python
data = {"fruits": ["apple","grape","orange"],"vegies":["potato"],"fruits":"test"}

data

```

```python
data = {"fruits","vegies"}
```

```python
data = {"fruit": "apple","vegies":"potato"}

data

```

Dict support multiple data types as a value

```python
data = {"fruit": 1,"vegies":"potato","numbers":[],"test":(),"test2":{},"test3":False,"test4": {"d":"data"}}
data

```

Data assigning

```python
# data[0] = "apple"
data["a"] = []
data
```

```python
data["a"] = "udhaya"
data
```

Dict creation using dict class

```python
dic = dict(a=[],b=[])
dic["c"] = []
dic
```

```python
# accessing an dictionary item
print(dic["a"])
```

```python
dic['a'] = "udhaya"
dic
```

dict get method

```python
dic.get("a")
```

```python
dic["z"]
```

```python
dic.get("z",False)
```

Looping the dict

```python
for data in dic:
    print(data)
```

```python
for data in dic.items():
    print(data)
```

```python
for key,value in dic.items():
    print(key,value)
```

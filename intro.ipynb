```python
### 9️⃣ Compress a String

# Write a function that performs basic string compression using character counts (like run-length encoding).

# **Example:**
# Input: `"aaabbc"`
# Output: `"a3b2c1"`


# aaaaabbbbbccccc

# a5b5c5



value = "aaaaabbbbbccccc"
# value = "aaabbc"
# a5b5c5

ans = ""
i = 0
n = len(value)

while i < n:
    print(f"i: ",i)
    count = 1
    j = i
    print("outer j: ",j)
    while j + 1 < n and value[j] == value[j+1]:
        print(f"value[{j}] = ",value[j], f"value[{j+1}] = ",value[j+1])
        j += 1
        count += 1
    print("count: ",count)

    ans = ans + value[i]+ str(count) 
    print(ans)
    i = j + 1
    print("final i: ",i)

print(ans)
```

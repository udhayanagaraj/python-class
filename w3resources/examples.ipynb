18. Palindrome Filter Lambda

Write a Python program to find palindromes in a given list of strings using Lambda.

Orginal list of strings:
['php', 'w3r', 'Python', 'abcd', 'Java', 'aaa']
List of palindromes:
['php', 'aaa']

```python
def is_palindrome(list_arr):
    ans = []

    for i in list_arr:
        if i[::-1] == i:
            ans.append(i)
    
    return ans

answer = is_palindrome(['php', 'w3r', 'Python', 'abcd', 'Java', 'aaa'])
print(answer)
def is_greater_three(list_arr,k):
    ans = []

    for i in list_arr:
        if len(i) > k:
            ans.append(i)
    
    return ans

final = is_greater_three(answer,3)
       
final
```
